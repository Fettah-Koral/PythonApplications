```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=0).fit(std_veri.iloc[:,1:])
```

```python
kmeans.labels_
```

```python
from collections import Counter
Counter(kmeans.labels_)
```

```python
kmeans.cluster_centers_
```

```python
kume_listesi=pd.Series(data=kmeans.labels_)
kume_listesi
```

```python
kumelenmis_veri=std_veri.copy()
```

```python
kumelenmis_veri["Kumesi"]=kume_listesi
```

```python
kmeans.labels_
```
