```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3,4],[1,4,9,16]);
```

```python
x =[1,2,3,4]
y =[1,4,9,16]

plt.xlabel("x_axis")
plt.ylabel("y_axis")
plt.title("ilk grafiğimiz ")
plt.plot(x,y,label="veri etiketi",color="seagreen")
plt.legend();
```

```python
plt.xticks([1,2,3,4])
plt.yticks([1,4,9,16])
plt.xlabel("x_axis")
plt.ylabel("y_axis")
plt.title("ilk grafiğimiz ")
plt.plot(x,y,label="X^2",color="b",marker="H",linestyle="--",linewidth=1)
plt.legend();
```

```python
plt.xticks([1,2,3,4])
plt.yticks([1,4,9,16])
plt.xlabel("x_axis")
plt.ylabel("y_axis")
plt.title("ilk grafiğimiz ")
plt.plot(x,y,label="X ^ 2",color="k",marker="H",linestyle="--",linewidth=2)
plt.legend();

x2=np.arange(0,5,0.5)
plt.plot(x2,x2*2,"m+",linewidth=3,linestyle="-.",label="X * 2")
plt.legend()
plt.savefig("3.png")
plt.show()
```

```python
data = pd.read_csv("PythonProgramlamaVeYapayZeka-main/petrolfiyatlari.csv")
data.tail(2)
```

```python
plt.title("petrol prices")
plt.plot(data["Year"],data["USA"],"b-",label="USA")

plt.xlabel("Year")
plt.ylabel("Dolar")

plt.legend()
plt.show()
```

```python
plt.title("petrol prices")
plt.plot(data["Year"],data["USA"],"b-",label="USA")
plt.plot(data["Year"],data["Canada"],"g.-",label="Canada")
plt.plot(data["Year"],data["Japan"],"r-",label="Japan")
plt.plot(data["Year"],data["UK"],marker="*",color="c",label="UK")
plt.plot(data["Year"],data["Germany"],marker="x",color="m",label="Germany")

plt.xlabel("Year")
plt.ylabel("Dolar")

plt.legend()
plt.savefig("4.png")
plt.show()
```
