```python
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
df = pd.read_csv("polynomial.csv",sep=";")
df.tail(1)
```

```python
plt.scatter(df["deneyim"],df["maas"])
plt.xlabel("Deneyim")
plt.ylabel("Maas")
plt.savefig('1.png',dpi=300)
plt.show()
```

```python
reg = LinearRegression()
reg.fit(df[["deneyim"]],df.maas)
```

```python
plt.scatter(df.deneyim,df.maas)
plt.plot(df.deneyim,reg.predict(df[["deneyim"]]),color="red",label="linear regression")
plt.legend()
plt.xlabel("deneyim")
plt.ylabel("maaş")
plt.savefig("2.png")
plt.show()
```

```python
polynomial_regression=PolynomialFeatures(degree = 4)
x_polynomial = polynomial_regression.fit_transform(df[["deneyim"]])
```

```python
reg = LinearRegression()
reg.fit(x_polynomial,df["maas"])
```

```python
x_polynomial
```

```python
reg.predict(x_polynomial)
```

```python
y_head = reg.predict(x_polynomial)
plt.plot(df["deneyim"],y_head,color="purple",label="polynomial regression")
plt.scatter(df["deneyim"],df["maas"])
plt.legend()
plt.show()
```
